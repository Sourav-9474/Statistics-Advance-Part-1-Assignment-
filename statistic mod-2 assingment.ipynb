**Statistics Advance Part 1- Assignment Questions**

 1. What is a random variable in probability theory?

 => A random variable is a variable that takes on different numerical values based on the outcomes of a random phenomenon. It maps outcomes from a sample space to real numbers.

2.  What are the types of random variables?

=>  There are two main types:

Discrete Random Variable: Takes on a countable number of values (e.g., number of heads in 3 coin flips).

Continuous Random Variable: Takes on an infinite number of possible values within a range (e.g., height, weight, time).

3.  What is the difference between discrete and continuous distributions?

=>  Discrete Distribution: Deals with discrete random variables and assigns probabilities to specific values.

Continuous Distribution: Deals with continuous random variables, and probabilities are represented over intervals using a probability density function (PDF).

4.  What are probability distribution functions (PDF)?

=>  For continuous variables, a PDF describes the relative likelihood of a random variable to take on a given value. The area under the PDF curve over an interval gives the probability that the variable falls within that interval.

5.  How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?

=>  PDF: Shows the likelihood (density) at each point.

CDF: Shows the cumulative probability up to a certain value (i.e., P(X ≤ x)). It is the integral of the PDF.



6. What is a discrete uniform distribution?

=> A discrete uniform distribution is one where all outcomes have an equal probability. Example: Rolling a fair die (each side has probability 1/6).

7. What are the key properties of a Bernoulli distribution/

=>  Two outcomes: success (1) and failure (0).

One trial.

Defined by a single parameter p = probability of success.

Mean = p; Variance = p(1 - p).



8.  What is the binomial distribution, and how is it used in probability?

=>  Represents the number of successes in n independent Bernoulli trials.

Parameters: n (number of trials), p (probability of success).

Used in modeling success/failure experiments (e.g., number of heads in 10 coin tosses).



9.  What is the Poisson distribution and where is it applied?

=>  Models the number of events occurring in a fixed interval of time/space.

Assumes events occur independently at a constant average rate λ.

Used in scenarios like customer arrivals, phone calls per hour, etc.



10. What is a continuous uniform distribution?

=>  A distribution where every value within a certain range is equally likely. Example: Randomly choosing a number between 0 and 1.

11.  What are the characteristics of a normal distribution?

=>  Bell-shaped, symmetric around the mean.

Defined by two parameters: mean (μ) and standard deviation (σ).

Total area under the curve = 1.

Commonly arises in natural and social phenomena.

12.  What is the standard normal distribution, and why is it important?

=> A special case of the normal distribution with mean = 0 and standard deviation = 1.

Important for standardizing variables and calculating probabilities using Z-scores.



13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?

=>  CLT states that the sampling distribution of the sample mean approaches a normal distribution as the sample size becomes large, regardless of the population’s distribution. It's critical because it enables:

Use of normal approximation.

Confidence interval estimation.

Hypothesis testing.

14.  How does the Central Limit Theorem relate to the normal distribution?

=>  CLT explains why the normal distribution appears so frequently in statistics — it describes the distribution of sample means, not just raw data.

15. What is the application of Z statistics in hypothesis testing?

=>  Z-statistics are used to determine how far a sample statistic is from the population parameter, in terms of standard errors. They're crucial in testing:

Mean differences.

Proportions.

Confidence intervals.

16.  How do you calculate a Z-score, and what does it represent?

=>  Z = (X - μ) / σ

It tells how many standard deviations a data point (X) is from the mean (μ). A Z-score:

Positive: above the mean.

Negative: below the mean.

Helps identify outliers and calculate probabilities.

17.  What are point estimates and interval estimates in statistics?

=>  Point Estimate: A single value estimate of a population parameter (e.g., sample mean).

Interval Estimate: A range of values (confidence interval) that likely includes the population parameter.

18.   What is the significance of confidence intervals in statistical analysis?

=>  Confidence intervals express the range of plausible values for an unknown parameter and reflect the uncertainty of point estimates. Common confidence levels are 90%, 95%, and 99%.



19.   What is the relationship between a Z-score and a confidence interval?

=>  Confidence intervals are often constructed using Z-scores (especially with known variance or large sample sizes):

CI = sample mean ± Z * (standard error)

The Z-value corresponds to the desired confidence level (e.g., 1.96 for 95%).



20.  How are Z-scores used to compare different distributions?

=>  Z-scores standardize data from different distributions, allowing comparisons on a common scale. Useful when the original distributions have different means or variances.



21.   What are the assumptions for applying the Central Limit Theorem?

=>  Independent samples.

Identically distributed (same distribution).

Sample size should be large enough (n ≥ 30 is a good rule of thumb).

For small samples, population should be approximately normal.

22.  What is the concept of expected value in a probability distribution?

=>  The expected value (E[X]) is the mean or average outcome you'd expect if an experiment is repeated many times:

Discrete: E[X] = Σx * P(x)

Continuous: E[X] = ∫x * f(x) dx

23.  How does a probability distribution relate to the expected outcome of a random variable?

=> The probability distribution defines how likely each outcome is, and the expected value is the weighted average of those outcomes based on their probabilities.

```python
#14.  Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution?
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

def calculate_z_score(x, mean, std):
    z = (x - mean) / std
    return z

def compare_to_standard_normal(z):
    # Get cumulative probability for Z
    prob = norm.cdf(z)
    print(f"Z-score: {z:.2f}")
    print(f"Cumulative Probability (P(Z ≤ {z:.2f})): {prob:.4f}")

    # Plot the standard normal distribution and shade the area up to the Z-score
    x_vals = np.linspace(-4, 4, 1000)
    y_vals = norm.pdf(x_vals)

    plt.figure(figsize=(10, 5))
    plt.plot(x_vals, y_vals, label='Standard Normal PDF', color='blue')
    plt.fill_between(x_vals, y_vals, 0, where=(x_vals <= z), color='skyblue', alpha=0.6, label=f'P(Z ≤ {z:.2f})')
    plt.axvline(z, color='red', linestyle='--', label=f'Z = {z:.2f}')
    plt.title('Z-score on the Standard Normal Distribution')
    plt.xlabel('Z')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage
x = 75
mean = 70
std = 10

z = calculate_z_score(x, mean, std)
compare_to_standard_normal(z)
```

```python
#15.  Implement hypothesis testing using Z-statistics for a sample dataset?
import numpy as np
from scipy.stats import norm

def z_test(sample_data, population_mean, population_std, alpha=0.05):
    n = len(sample_data)
    sample_mean = np.mean(sample_data)

    # Z-statistic
    z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(n))

    # Critical Z for two-tailed test
    z_critical = norm.ppf(1 - alpha / 2)

    # P-value
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))

    print(f"Sample Mean = {sample_mean:.2f}")
    print(f"Z-statistic = {z_stat:.4f}")
    print(f"Critical Z-value = ±{z_critical:.4f}")
    print(f"P-value = {p_value:.4f}")

    if abs(z_stat) > z_critical:
        print("Reject the null hypothesis (significant difference).")
    else:
        print("Fail to reject the null hypothesis (no significant difference).")

# Example: test if sample mean differs from 100
np.random.seed(42)
sample = np.random.normal(loc=102, scale=15, size=30)
z_test(sample_data=sample, population_mean=100, population_std=15, alpha=0.05)
```

```python
#16  Create a confidence interval for a dataset using Python and interpret the result?

import numpy as np
from scipy.stats import norm

def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    std_dev = np.std(data, ddof=1)  # sample std deviation
    z_critical = norm.ppf(1 - (1 - confidence) / 2)

    margin_of_error = z_critical * (std_dev / np.sqrt(n))

    ci_lower = mean - margin_of_error
    ci_upper = mean + margin_of_error

    print(f"Sample Mean: {mean:.2f}")
    print(f"{int(confidence*100)}% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")

    return (ci_lower, ci_upper)

# Example dataset
np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=8, size=40)

# Calculate and interpret the confidence interval
confidence_interval(sample_data, confidence=0.95)
```

```python
#17.  Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean?
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def generate_data_and_confidence_interval(mean=100, std=15, size=50, confidence=0.95):
    # Step 1: Generate normal data
    np.random.seed(42)
    data = np.random.normal(loc=mean, scale=std, size=size)

    # Step 2: Compute sample mean and standard deviation
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)

    # Step 3: Compute margin of error and confidence interval
    z_critical = norm.ppf(1 - (1 - confidence) / 2)
    margin_error = z_critical * (sample_std / np.sqrt(size))

    ci_lower = sample_mean - margin_error
    ci_upper = sample_mean + margin_error

    # Step 4: Print and interpret
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"{int(confidence*100)}% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
    print(f"We are {int(confidence*100)}% confident the true population mean lies within this range.")

    # Step 5: Optional Plot
    plt.figure(figsize=(8, 4))
    plt.errorbar(1, sample_mean, yerr=margin_error, fmt='o', capsize=5, color='green', label='Confidence Interval')
    plt.axhline(sample_mean, color='blue', linestyle='--', label='Sample Mean')
    plt.xticks([])
    plt.title('Confidence Interval for the Mean of Normally Distributed Data')
    plt.ylabel('Value')
    plt.legend()
    plt.grid(True)
    plt.show()

# Run it
generate_data_and_confidence_interval()
```

```python
#18. Write a Python script to calculate and visualize the probability density function (PDF) of a normal distribution/
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_normal_pdf(mean=0, std=1):
    # Create a range of x values
    x = np.linspace(mean - 4*std, mean + 4*std, 1000)

    # Calculate the PDF using scipy
    pdf = norm.pdf(x, loc=mean, scale=std)

    # Plotting
    plt.figure(figsize=(10, 5))
    plt.plot(x, pdf, color='blue', label=f'Normal PDF\nMean={mean}, Std={std}')
    plt.fill_between(x, pdf, color='skyblue', alpha=0.4)
    plt.title('Probability Density Function of a Normal Distribution')
    plt.xlabel('x')
    plt.ylabel('Probability Density')
    plt.grid(True)
    plt.legend()
    plt.show()

# Example usage
plot_normal_pdf(mean=100, std=15)
```

```python
# 19  Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution?
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

def poisson_cdf_analysis(lam=4, max_k=15):
    # Generate range of x values (discrete counts)
    x = np.arange(0, max_k + 1)

    # Calculate Poisson CDF values
    cdf = poisson.cdf(x, mu=lam)

    # Print CDF values
    print(f"Poisson CDF (λ = {lam}):")
    for k, p in zip(x, cdf):
        print(f"P(X ≤ {k}) = {p:.4f}")

    # Plot the CDF
    plt.figure(figsize=(10, 5))
    plt.step(x, cdf, where='post', color='green', label='Poisson CDF')
    plt.scatter(x, cdf, color='black')
    plt.title(f'Cumulative Distribution Function of Poisson (λ = {lam})')
    plt.xlabel('k (number of events)')
    plt.ylabel('P(X ≤ k)')
    plt.ylim(0, 1.05)
    plt.grid(True)
    plt.legend()
    plt.show()

# Example usage
poisson_cdf_analysis(lam=4)
```

```python
# 21 Write a Python program to compare the standard deviations of two datasets and visualize the difference?
import numpy as np
import matplotlib.pyplot as plt

def compare_std_devs(size=1000):
    # Generate two datasets
    np.random.seed(42)
    data1 = np.random.normal(loc=50, scale=5, size=size)   # std dev = 5
    data2 = np.random.normal(loc=50, scale=15, size=size)  # std dev = 15

    # Calculate standard deviations
    std1 = np.std(data1, ddof=1)
    std2 = np.std(data2, ddof=1)

    print(f"Standard Deviation of Dataset 1: {std1:.2f}")
    print(f"Standard Deviation of Dataset 2: {std2:.2f}")

    # Plot histograms
    plt.figure(figsize=(10, 5))
    plt.hist(data1, bins=30, alpha=0.6, label='Dataset 1 (std=5)', color='blue', density=True)
    plt.hist(data2, bins=30, alpha=0.6, label='Dataset 2 (std=15)', color='orange', density=True)
    plt.title('Comparison of Standard Deviations')
    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.legend()
    plt.grid(True)
    plt.show()


compare_std_devs()
```

```python
# 22.  Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution?
import numpy as np
import matplotlib.pyplot as plt

def calculate_range_and_iqr(mean=50, std=10, size=1000):
    # Step 1: Generate normal data
    np.random.seed(42)
    data = np.random.normal(loc=mean, scale=std, size=size)

    # Step 2: Calculate range
    data_range = np.max(data) - np.min(data)

    # Step 3: Calculate IQR (Q3 - Q1)
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    iqr = q3 - q1

    print(f"Range: {data_range:.2f}")
    print(f"IQR (Interquartile Range): {iqr:.2f}")
    print(f"Q1: {q1:.2f}, Q3: {q3:.2f}")

    # Step 4: Visualize with a boxplot
    plt.figure(figsize=(6, 4))
    plt.boxplot(data, vert=False, patch_artist=True,
                boxprops=dict(facecolor='skyblue', color='blue'),
                medianprops=dict(color='red'),
                whiskerprops=dict(color='black'),
                capprops=dict(color='black'))
    plt.title('Boxplot of Normally Distributed Data')
    plt.xlabel('Value')
    plt.grid(True)
    plt.show()


calculate_range_and_iqr()
```

```python
# 23.  Implement Z-score normalization on a dataset and visualize its transformation?
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore

def z_score_normalization(size=1000):
    # Step 1: Generate a dataset (e.g., normal distribution)
    np.random.seed(42)
    data = np.random.normal(loc=100, scale=20, size=size)

    # Step 2: Apply Z-score normalization
    normalized_data = zscore(data)

    # Step 3: Print summary stats
    print(f"Original Data: Mean = {np.mean(data):.2f}, Std = {np.std(data):.2f}")
    print(f"Normalized Data: Mean = {np.mean(normalized_data):.2f}, Std = {np.std(normalized_data):.2f}")

    # Step 4: Visualize before and after
    plt.figure(figsize=(12, 5))

    # Original data
    plt.subplot(1, 2, 1)
    plt.hist(data, bins=30, color='skyblue', edgecolor='black', density=True)
    plt.title('Original Data')
    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.axvline(np.mean(data), color='red', linestyle='--', label='Mean')
    plt.legend()

    # Normalized data
    plt.subplot(1, 2, 2)
    plt.hist(normalized_data, bins=30, color='lightgreen', edgecolor='black', density=True)
    plt.title('Z-score Normalized Data')
    plt.xlabel('Z-score')
    plt.ylabel('Density')
    plt.axvline(0, color='red', linestyle='--', label='Mean (0)')
    plt.legend()

    plt.tight_layout()
    plt.show()

z_score_normalization()
```

```python
# 24.  Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution?
import numpy as np
from scipy.stats import skew, kurtosis

def analyze_skewness_kurtosis(mean=0, std=1, size=1000):
    # Generate normal distribution data
    np.random.seed(42)
    data = np.random.normal(loc=mean, scale=std, size=size)

    # Calculate skewness and kurtosis
    skewness = skew(data)
    kurt = kurtosis(data, fisher=True)  # Fisher=True means excess kurtosis (0 for normal)

    # Print the results
    print(f"Skewness: {skewness:.4f}")
    print(f"Kurtosis (excess): {kurt:.4f}")

    # Interpretation
    print("\nInterpretation:")
    if abs(skewness) < 0.5:
        print("- The data is approximately symmetric.")
    elif skewness > 0:
        print("- The data is positively skewed (right-tailed).")
    else:
        print("- The data is negatively skewed (left-tailed).")

    if abs(kurt) < 1:
        print("- Kurtosis indicates the data has a normal-like peak.")
    elif kurt > 0:
        print("- Leptokurtic: Sharper peak and heavier tails.")
    else:
        print("- Platykurtic: Flatter peak and lighter tails.")

analyze_skewness_kurtosis()
```
